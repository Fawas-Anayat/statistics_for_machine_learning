### Making graphs of the Statistical Functions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics as sts
```

```python
data={
    'scores':np.random.randint(10,100,30)
}
df=pd.DataFrame(data)
print(df['scores'])
(df['scores']).value_counts()
```

```python
medain=np.median(df['scores'])
mean=np.mean(df['scores'])
mod=sts.mode(df['scores'])
std=np.std(df['scores'])
var=np.var(df['scores'])
q3=np.quantile(df['scores'],[0.75])
print(medain)
print(mean)
print(mod)
print(std)
print(var)
print(q3)
```

```python
import seaborn as sns
from scipy.stats import norm

# -------------------------
# 1. DISCRETE CASE (PMF & CDF)
# -------------------------
die_faces = np.arange(1, 7)  # outcomes of a fair die
pmf_values = np.ones(6) / 6  # equal probability for each face

# CDF for discrete distribution
cdf_values = np.cumsum(pmf_values)

# -------------------------
# 2. CONTINUOUS CASE (PDF & CDF)
# -------------------------
x = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(x, loc=0, scale=1)   # standard normal distribution
cdf_continuous = norm.cdf(x, loc=0, scale=1)

# -------------------------
# PLOTTING
# -------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot PMF
axes[0].bar(die_faces, pmf_values, color='skyblue', edgecolor='black')
axes[0].set_title("PMF (Discrete) - Fair Die")
axes[0].set_xlabel("Die Face")
axes[0].set_ylabel("Probability")

# Plot Discrete CDF
axes[1].step(die_faces, cdf_values, where='mid', color='green')
axes[1].set_title("CDF (Discrete) - Fair Die")
axes[1].set_xlabel("Die Face")
axes[1].set_ylabel("Cumulative Probability")

# Plot PDF & CDF (Continuous)
axes[2].plot(x, pdf_values, label="PDF (Normal)", color='blue')
axes[2].plot(x, cdf_continuous, label="CDF (Normal)", color='red')
axes[2].set_title("PDF & CDF (Continuous Normal)")
axes[2].set_xlabel("x")
axes[2].set_ylabel("Density / Probability")
axes[2].legend()

plt.tight_layout()
plt.show()
```

### Normal Distribution

```python
# Mean and standard deviation
mu, sigma = 0, 1

# Generate x values
x = np.linspace(-4, 4, 1000)

# PDF of normal distribution
pdf = norm.pdf(x, mu, sigma)

# CDF of normal distribution
cdf = norm.cdf(x, mu, sigma)

plt.figure(figsize=(10,6))

# Plot PDF
plt.plot(x, pdf, label="PDF (Density)")
plt.fill_between(x, pdf, alpha=0.2)

# Plot CDF
plt.plot(x, cdf, label="CDF", linestyle="--")

plt.title("Normal Distribution (PDF & CDF)")
plt.xlabel("x")
plt.ylabel("Probability / Density")
plt.legend()
plt.grid(True)
plt.show()
```

### Standard Normal Distribution

```python
# Standard normal distribution parameters
mu = 0      # mean
sigma = 1   # standard deviation

# Generate x values
x = np.linspace(-4, 4, 1000)  # range for standard normal
y = norm.pdf(x, mu, sigma)    # probability density function (PDF)

# Plot the normal distribution curve
plt.plot(x, y, label="Standard Normal Distribution", color="blue")

# Highlight mean
plt.axvline(mu, color="red", linestyle="--", linewidth=2, label="Mean ( = 0)")

# Highlight 1 and 2
for i in range(1, 3):
    plt.axvline(mu + i*sigma, color="green", linestyle="--", linewidth=1.5, label=f"+{i}")
    plt.axvline(mu - i*sigma, color="orange", linestyle="--", linewidth=1.5, label=f"-{i}")

# Titles and labels
plt.title("Standard Normal Distribution (=0, =1)")
plt.xlabel("x values")
plt.ylabel("Density (PDF)")
plt.legend(loc="upper left")

plt.grid(True)
plt.show()
```

### Bernoulli Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Probability of success (p)
p = 0.3  

# Possible outcomes: 0 (failure), 1 (success)
x = [0, 1]

# PMF (Probability Mass Function)
pmf = bernoulli.pmf(x, p)

# CDF (Cumulative Distribution Function)
cdf = bernoulli.cdf(x, p)

# --- Plot PMF ---
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.stem(x, pmf,basefmt=" ")
plt.title(f"Bernoulli PMF (p={p})")
plt.xlabel("x (Outcome)")
plt.ylabel("P(X=x)")
plt.xticks([0,1])

# --- Plot CDF ---
plt.subplot(1,2,2)
plt.step(x, cdf, where="post")
plt.title(f"Bernoulli CDF (p={p})")
plt.xlabel("x (Outcome)")
plt.ylabel("P(X  x)")
plt.xticks([0,1])

plt.tight_layout()
plt.show()
```

### Uniform Distribution

```python
from scipy.stats import uniform

# Parameters
a, b = 0, 10   # range of uniform distribution

# PDF
x = np.linspace(-2, 12, 1000)
pdf = uniform.pdf(x, loc=a, scale=b-a)

# CDF
cdf = uniform.cdf(x, loc=a, scale=b-a)

# Plot
plt.figure(figsize=(12,5))

# PDF
plt.subplot(1,2,1)
plt.plot(x, pdf, 'b-', lw=2, label='PDF')
plt.fill_between(x, pdf, alpha=0.2, color='blue')
plt.title('Uniform Distribution PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()

# CDF
plt.subplot(1,2,2)
plt.plot(x, cdf, 'g-', lw=2, label='CDF')
plt.title('Uniform Distribution CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.show()
```
